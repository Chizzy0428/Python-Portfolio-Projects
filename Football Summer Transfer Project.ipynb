```python
#average age of the players
average_age = df['Age'].mean()
print('The average age of the players is:', average_age)
```

```python
#Which positions are most common among the players?
position_counts = df['Position'].value_counts()
position_counts = position_counts.to_frame()
print('Positions and their counts:')
position_counts
#The CB position was the most sought after position in the transfer window

```

```python
#Which countries has the most incomings in the dataset?
country_counts = df['Country_New_Club'].value_counts()
country_counts  = country_counts .to_frame()
print('Countries With the highest Incomings:')
country_counts.head()
#Italy is the country with the highest incommings
```

```python
#Which countries has the most outgoings in the dataset?
country_counts2 = df['Country_Old_Club'].value_counts()
country_counts2  = country_counts2.to_frame()
print('Countries With the highest Outgoings:')
country_counts2.head()
#Italy is the country with the highest Outgoings

```

```python
#Are there any notable transfers with exceptionally high or low transfer fees?
highest_transfers = df.sort_values('Transfer_Fee', ascending=False)
print("Transfers with the highest fees:")
highest_transfers= highest_transfers.reset_index(drop=True)
pd.DataFrame(highest_transfers.head(5))
```

```python
#Are there any notable transfers with exceptionally high or low transfer fees?
lowest_transfers = df.sort_values('Transfer_Fee',  ascending=True)
print("Transfers with the Lowest fees:")
lowest_transfers = lowest_transfers.reset_index(drop=True)
pd.DataFrame(lowest_transfers.head(5))
```

```python
 #Is there any correlation between player age and transfer fee?
# Remove unwanted characters from 'Transfer_Fee' column
df['Transfer_Fee'] = df['Transfer_Fee'].str.replace('[^\d.]', '', regex=True)
#convert to numeric
df['Transfer_Fee'] = pd.to_numeric(df['Transfer_Fee'], errors='coerce')
correlation = df['Age'].corr(df['Transfer_Fee'])
print('Correlation coefficient:', correlation)
#There is no Correlation
```

```python
#Is there any correlation between player value and transfer fee?
df['Player_Value'] = df['Player_Value'].str.replace('[^\d.]', '', regex=True)
#convert to numeric
df['Player_Value'] = pd.to_numeric(df['Player_Value'], errors='coerce')
correlation2 = df['Player_Value'].corr(df['Transfer_Fee'])
print('Correlation coefficient:', correlation2)
```

```python
# Create a scatter plot
plt.scatter('Player_Value', 'Transfer_Fee')
plt.title('Player Value vs Transfer Fee');
```

This Correlation coefficient: 0.3170557439195942 and the Scatter Plot tells us that there is a positive correlation between player age and transfer fee but, the correlation is not very strong.
